```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Membaca dataset menus
menus = pd.read_csv('./recommendation/menus.csv')
menus
```

```python
# Membaca dataset ratings
ratings = pd.read_csv('./recommendation/ratings.csv')
ratings
```

```python
# Menggabungkan dataset
data = pd.merge(menus, ratings)
data
```

```python
# Drop data yang tidak diperlukan
data.drop(["comment", "id"], axis=1, inplace=True)
data
```

```python
data_pivot = data.pivot_table(index="userId", columns="menu", values="rating")
data_pivot
```

```python
# Make dataset for the second way
data_pivot2 = pd.pivot(index="menuId", columns="userId", data=ratings, values="rating")
data_pivot2
```

```python
menu_voted = pd.DataFrame(ratings.groupby("menuId")["rating"].agg("count"))
menu_voted.reset_index(level=0, inplace=True)
menu_voted
```

```python
user_voted = pd.DataFrame(ratings.groupby("userId")["rating"].agg("count"))
user_voted.reset_index(level = 0, inplace = True)
user_voted
```

```python
data_pivot2.fillna(0, inplace=True)
data_pivot2
```

```python
# Last version dataset yang akan digunakan
data_last = data_pivot2.loc[menu_voted[menu_voted["rating"] > 10]["menuId"], :]
data_last = data_last.loc[:, user_voted[user_voted["rating"] > 60]["userId"]]
data_last
```

```python
# Recommendation system
from scipy.sparse import csr_matrix
csr_data = csr_matrix(data_last.values)
data_last.reset_index(inplace=True)
```

```python
csr_data
```

```python
from sklearn.neighbors import NearestNeighbors
knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=20, n_jobs=-1)
knn.fit(csr_data)
```

```python
def menu_recommendation(menu_name):
    
    n_menus_to_reccomend = 10
    menu_list = menus[menus['menu'].str.contains(menu_name)] # Find movie name
    
    if len(menu_list):
        
        menu_idx = menu_list.iloc[0]['menuId']
        menu_idx = data_last[data_last['menuId'] == menu_idx].index[0]
        
        distances, indices = knn.kneighbors(csr_data[menu_idx],n_neighbors=n_menus_to_reccomend+1)    
        rec_menu_indices = sorted(list(zip(indices.squeeze().tolist(),distances.squeeze().tolist())),key=lambda x: x[1])[:0:-1]
        recommend_frame = []
        
        for val in rec_menu_indices:
            menu_idx = data_last.iloc[val[0]]['menuId']
            idx = menus[menus['menuId'] == menu_idx].index
            recommend_frame.append({'Title':menus.iloc[idx]['menu'].values[0],'Distance':val[1]})
        df = pd.DataFrame(recommend_frame,index=range(1,n_menus_to_reccomend+1))
        return df
    
    else:
        return "No menu found. Please check your input"
```

```python
# result of Recommendation System
menu_recommendation('bakso')
```
